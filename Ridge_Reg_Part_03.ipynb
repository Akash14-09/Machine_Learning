#Ridge Reg on Gradient Descent

```python
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score
import numpy as np
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
reg = SGDRegressor(penalty = '12' , max_iter = 500 , eta0 = 0.1 , learning_rate = 'constant' , alpha = 0.001)
```

```python
class MyRidge:
  def __init__(self , epochs , learning_rate , alpha):
    self.epochs = epochs
    self.learning_rate = learning_rate
    self.alpha = alpha
    self.weights = None
    self.bias = None

  def fit(self , X_train , y_train):
    self.coef = np.ones(X_train.shape[1])
    self.intercept = 0
    theta = np.insert(self.coef , 0 , self.intercept)
    X_train = np.insert(X_train , 0 , 1 , axis = 1)
    for i in range(self.epochs):
      theta_der = np.dot(X_train.T , X_train).dot(theta) - np.dot(X_train.T , y_train) + self.alpha * theta
      theta = theta - self.learning_rate * theta_der
    self.intercept = theta[0]
    self.coef_ = theta[1:]

  def predict(self , X_test):
    return np.dot(X_test , self.coef_) + self.intercept
```

```python
reg = MyRidge(epochs = 500 , learning_rate = 0.005 , alpha = 0.001)
reg.fit(X_train , y_train)
y_pred = reg.predict(X_test)
```

```python
print("R2 score",r2_score(y_test , y_pred))
print(reg.coef_)
print(reg.intercept)
```
