##Bagging Regressor

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score
```

```python
housing = fetch_california_housing()

X = housing.data
y = housing.target

print("Feature Names:")
print(housing.feature_names)

print("\nFeature Shape:", X.shape)
print("Target Shape:", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    train_size=0.80,
    test_size=0.20,
    random_state=123
)
print("\nTrain/Test Shapes:")
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python

lr = LinearRegression()
dt = DecisionTreeRegressor(random_state=42)
knn = KNeighborsRegressor()
```

```python

lr.fit(X_train, y_train)
dt.fit(X_train, y_train)
knn.fit(X_train, y_train)
```

```python

y_pred_lr = lr.predict(X_test)
y_pred_dt = dt.predict(X_test)
y_pred_knn = knn.predict(X_test)
```

```python

print("\nR^2 Scores")

print("Linear Regression:",
      r2_score(y_test, y_pred_lr))

print("Decision Tree:",
      r2_score(y_test, y_pred_dt))

print("KNN:",
      r2_score(y_test, y_pred_knn))
```

```python
bag_regressor = BaggingRegressor(
    estimator=DecisionTreeRegressor(),
    n_estimators=10,
    random_state=1
)

bag_regressor.fit(X_train, y_train)
```

```python
y_pred_bag = bag_regressor.predict(X_test)
```

```python

print("\nBagging Regressor Scores")

print("Training R^2 Score:",
      bag_regressor.score(X_train, y_train))

print("Testing R^2 Score:",
      bag_regressor.score(X_test, y_test))
```

```python
params = {
    'estimator': [
        None,
        LinearRegression(),
        KNeighborsRegressor()
    ],

    'n_estimators': [20, 50],

    'max_samples': [0.5, 1.0],

    'max_features': [0.5, 1.0],

    'bootstrap': [True, False],

    'bootstrap_features': [True, False]
}
```

```python
grid = GridSearchCV(
    estimator=BaggingRegressor(
        random_state=1,
        n_jobs=-1
    ),

    param_grid=params,

    cv=3,

    n_jobs=-1,

    verbose=1
)
```

```python
grid.fit(X_train, y_train)
```

```python
print("\nBest Results From Grid Search")

print("Best Cross Validation Score:",
      grid.best_score_)

print("Best Parameters:")
print(grid.best_params_)

print("\nTrain R^2 Score:",
      grid.best_estimator_.score(X_train, y_train))

print("Test R^2 Score:",
      grid.best_estimator_.score(X_test, y_test))
```
