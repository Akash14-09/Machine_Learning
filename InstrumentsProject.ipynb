```python
import random
import time
import matplotlib.pyplot as plt
import pandas as pd



# FOR TEMPRATURE DETECTION

def read_temperature():

    #CALCULATING THE TEMP IN RANGE 25 - 35 DEGREE C
    true_temp = random.uniform(25, 35)

    # ADDING ERRORS
    systematic_error = 0.5                 # Constant bias error
    random_error = random.uniform(-0.3, 0.3)   # Random noise ±0.3°C

    measured_temp = true_temp + systematic_error + random_error


    return round(measured_temp, 2), round(true_temp, 2) #round is used to keep the values upto two decimal places





#CALCULATION

temperature_data = []
true_data = []

print("Collecting temperature data... (24 readings)")

# Collecting 24 readings hourly
for _ in range(24):
    temp, true_value = read_temperature()   # Reading values from the above function

    print("Measured:", temp, " | True:", true_value)

    temperature_data.append(temp)       # Store measured
    true_data.append(true_value)        # Store true value

    time.sleep(1)




#CONVERSION INTO DATAFRAME

df = pd.DataFrame({
    "True Temperature": true_data,
    "Measured Temperature": temperature_data
})

# Error calculations
df["Absolute Error"] = df["Measured Temperature"] - df["True Temperature"]
df["Relative Error (%)"] = (df["Absolute Error"] / df["True Temperature"]) * 100






#VISULISATION

print("\nTemperature Data (With Error):")
print(df)


plt.figure(figsize=(18, 8)) #defining the figure size
plt.plot(df["Measured Temperature"], marker='o' , color='red', label="Measured")#Plotting the measured temprature
plt.plot(df["True Temperature"], marker='x', label="True") #plotting the true temprature
plt.title("Temperature Data Logger") #title of our plot
plt.xlabel("Time (in hours)") #x axis - hours
plt.ylabel("Temperature (°C)") #y axis - temprature
# plt.grid(True)
plt.legend()
plt.show()
```
