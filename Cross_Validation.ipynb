###Why can we not rely on Train Test Split


```python
from sklearn.datasets import load_iris
import numpy as np
```

```python
data = load_iris()
X = data['data']
y = data['target']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size = 0.3 , random_state = 4)
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
accuracy_score(y_test,y_pred)
```

Here you can see the problem by urself that when you change value of random_state then the accruracy score changes significantly .
So one solution we can apply the loop wise values for the random_state but it wont be optimal for optimal we would be using KFold Cross Validation

##KFold Cross Validation

```python
from sklearn.model_selection import KFold
```

```python
X = np.random.random(size = (25 , 5))
y = np.random.randint(low = 0 , high = 2 , size = 25)
```

```python
X
```

```python
y
```

```python
kf = KFold(n_splits = 6)
for train_index , test_index in kf.split(X):
  print("Train Index : " , train_index , "Test Index : " , test_index)
```

##What if there are unqual proportions of clasees in the Test Set?

####cross_val_score(classification)

```python
from sklearn.model_selection import cross_val_score
```

```python
data = load_iris()
X = data['data']
y = data['target']
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)
scores = cross_val_score(knn , X , y , cv = 10 , scoring = 'accuracy')
print(scores)
```

```python
scores.mean()
```

```python
for i in range(1,21) :
  knn = KNeighborsClassifier(n_neighbors = i)
  scores = cross_val_score(knn , X , y , cv = 10 , scoring = 'accuracy')
  print(i , scores.mean())
```
