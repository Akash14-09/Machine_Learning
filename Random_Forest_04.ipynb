#Random Forest V/s Bagging


###The basic difference is this that the random forest uses node sampling before plotting the decision tree while the bagging does row/col sampling before plotting the decision tree . Thats why random forest is better.

```python


import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
X,y = make_classification(n_features=5, n_redundant=0, n_informative=5,n_clusters_per_class=1)
```

```python
df = pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target'] = y
print(df.shape)
df.head()
```

```python
bag = BaggingClassifier(max_features=2)
#doing column sampling take 2 col out of 5
```

```python
bag.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python

plt.figure(figsize=(12,12))
plot_tree(bag.estimators_[0])
plt.show()
```

```python
rf = RandomForestClassifier(max_features=2)
```

```python
rf.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
plot_tree(rf.estimators_[4])
plt.show()
```

####you can see in the image that more is the randomnness occuring in the random forest graph thats means it is much better as compared to bagging one.
