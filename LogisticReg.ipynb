```python
def perceptron(X,y):

  X = np.insert(X,0,1,axis = 1)
  weights = np.ones(X.shape[1])
  lr = 0.1

  for i in range(1000):
    j = np.random.randint(0,100)
    y_hat = step(np.dot(X[j],weights))
    weights = weights + lr*(y[j] - y_hat)*X[j]

  return weights[0],weights[1:]
```

```python
def step(z):
  return 1 if z>0 else 0
```

```python
intercept_ , coef = perceptron(X,y)
```

```python
def perceptron(X,y):

  m = []
  b = []
  X = np.insert(X,0,1,axis = 1)
  weights = np.ones(X.shape[1])
  lr = 0.1

  for i in range(200):
    j = np.random.randint(0,100)
    y_hat = step(np.dot(X[j],weights))
    weights = weights + lr*(y[j] - y_hat)*X[j]

    m.append(-(weights[1]/weights[2]))
    b.append(-(weights[0]/weights[2]))

  return m,b
```

```python
m,b = perceptron(X,y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

fig, ax = plt.subplots(figsize=(9, 5))

m_animation, b_animation = perceptron(X, y)

x_i = np.arange(-3, 3, 0.1)

ax.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', s=100)
line, = ax.plot(x_i, x_i * m_animation[0] + b_animation[0], 'r-', linewidth=2)
plt.ylim(-3, 3)

def update(i):
    label = 'epoch {0}'.format(i + 1)
    line.set_ydata(x_i * m_animation[i] + b_animation[i])
    ax.set_xlabel(label)
    return line,

anim = FuncAnimation(fig, update, frames=len(m_animation), interval=100, repeat=True)

HTML(anim.to_jshtml())
```
