In this imputation you create a seprate category(column) for the missing values as here we can see the name of the category is 'missing'

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv',usecols = ['GarageQual' , 'FireplaceQu' , 'SalePrice'])
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
df['GarageQual'].value_counts().sort_values(ascending = False).plot.bar()
plt.xlabel('GarageQual')
plt.ylabel('No of houses')
```

```python
df['GarageQual'].fillna('Missing',inplace = True)
```

```python
df['GarageQual'].value_counts().sort_values(ascending = False).plot.bar()
plt.xlabel('GarageQual')
plt.ylabel('No of houses')
```

Implementing the same logic using Slickit Learn

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(df.drop(columns = ['SalePrice']) , df['SalePrice'] , test_size = 0.2 , random_state = 0)
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy = 'constant' , fill_value = 'missing')
```

```python
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)
```
